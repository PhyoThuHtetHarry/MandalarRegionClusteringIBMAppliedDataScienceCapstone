```python
#Function to extract the data for Mandalay Region
def MandalarRegion(dataframe):
    mandalay =  dataframe[dataframe[dataframe.columns[1]]=='Mandalay'].reset_index(drop=True)
    mandalay.columns = [str.replace('-','_') for str in mandalay]
    mandalay.columns = [str.replace(' ','') for str in mandalay]
    return mandalay
```

Dealing with unkown and Null Values: There are many things that we have to consider in this case. Cause the data represents the township level, we have to take care about it. And we can not use 'mean value' cause the data is really weird. Instead of this, the thing would be to replace with zero.

```python
def dealingwithunknown(dataset):
    dataset = dataset
    for i in dataset.columns[6:]:
        dataset.loc[dataset[i] == 'unknown',i]= 0
        dataset.loc[dataset[i] == 'Unknown',i]= 0
        dataset.loc[dataset[i] == 'Unkown',i]= 0
        dataset.loc[dataset[i] == '32(IN)',i]= 0
    return dataset
```

```python
def dealingwithnullvalues(dataset):
    dataset = dataset
    for i in dataset.columns[6:]:
        replace = []
        data  = dataset[i].isnull()
        print(data)
        for j,k in zip(data,dataset[i]):
            if (j==True):
                replace.append(0)
            else:
                replace.append(k)
        dataset[i] = replace
    return dataset
```

# Evaluation of Clustering

Since there are no labels in the clustering cases, we need to use different approachs. We need to have
compact clusters and well separated clusters that will lead to good model. The summary stastics that we will


use is WCSS(Within Cluster Sum of Square), BCSS(Between Cluster Sum of Square) and sillohette.

In this scanerio I will only show for the first one (educaiton_cases).

```python
kmeans2 = KMeans(n_clusters = 2,random_state=0)
km2 = kmeans2.fit_predict(X)
kmeans3 = KMeans(n_clusters = 3,random_state=0)
km3 = kmeans3.fit_predict(X)
kmeans4 = KMeans(n_clusters = 4,random_state=0)
km4 = kmeans4.fit_predict(X)
kmeans5 = KMeans(n_clusters = 5,random_state=0)
km5 = kmeans5.fit_predict(X)
kmeans6 = KMeans(n_clusters = 6,random_state=0)
km6 = kmeans6.fit_predict(X)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#BCSS WCSS and 
km_models = [kmeans2, kmeans3, kmeans4, kmeans5, kmeans6]

def plot_WCSS_km(km_models, samples):
    fig, ax = plt.subplots(1, 2, figsize=(12,5))
    
    ## Plot WCSS
    wcss = [mod.inertia_ for mod in km_models]
    n_clusts = range(2,len(wcss) + 2)
    ax[0].bar(n_clusts, wcss,color = 'green')
    ax[0].set_xlabel('Number of clusters')
    ax[0].set_ylabel('WCSS')
    
    ## Plot BCSS
    ## Compute BCSS as TSS - WCSS 
    n_1 = (float(samples.shape[0]) * float(samples.shape[1])) - 1.0
    tss = n_1 * np.var(samples)
    bcss = [tss - x for x in wcss]
    ax[1].bar(n_clusts, bcss)
    ax[1].set_xlabel('Number of clusters')
    ax[1].set_ylabel('BCSS')
    plt.show()
    
```

```python
plot_WCSS_km(km_models,X)
```

```python
from sklearn.metrics import silhouette_score
assignment_list = [km2, km3, km4, km5,km6]

def plot_sillohette(samples, assignments, x_lab = 'Number of clusters'):
    silhouette = [silhouette_score(samples, a) for a in assignments]
    n_clusts = range(2, len(silhouette) + 2)
    plt.bar(n_clusts, silhouette,color = 'red')
    plt.xlabel(x_lab)
    plt.ylabel('SC')
    plt.show()

plot_sillohette(X, assignment_list)      
```

For the data of education, when analyze with BCSS and WCSS, Kmeans with 6 clusters have the lowest value in WCSS and highest value in BCSS.
However in analyzing with sillohette metrics, it is clear that the clusters with 2 and 4 have the most highest values.
